```python
import matplotlib.pyplot as plt

import seaborn as sns
import pandas as pd
import numpy as np
```

```python
melbourne_data = melbourne_data.drop('BuildingArea', axis=1)
melbourne_data = melbourne_data.drop('YearBuilt', axis=1)
melbourne_data = melbourne_data.drop('CouncilArea', axis=1)
```

```python
round(100*(melbourne_data.isnull().sum()/len(melbourne_data.index)), 2)
```

```python
plt.figure(figsize=(18,10))
plt.plot(melbourne_data['Price'], color ='orange')
plt.title("Line Graph")
plt.xlabel("Row Indexes")
plt.ylabel("AGE")
plt.show()
```

```python
melbourne_data['Address'].sort_values().unique()
```

```python
plt.figure(figsize=(22,8))
sns.distplot(melbourne_data['Price'], color = 'darkviolet')
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.boxplot(data = melbourne_data, palette='magma')
```

# iris dataset

```python
iris_df = pd.read_csv('https://raw.githubusercontent.com/ammishra08/MachineLearning/master/Datasets/iris.csv')
```

```python
plt.figure(figsize=(10,10))
sns.set_style('darkgrid')
plt.scatter(x = iris_df['sepal_length'], y = iris_df['sepal_width'], label = 'sepal_width', marker='x', c='r', s=40)
plt.scatter(x = iris_df['sepal_length'], y = iris_df['petal_length'], label = 'petal_length', marker='x', s=40)
plt.xlabel('Sepal Length')
plt.legend(loc = 4)
plt.show()
```

```python
plt.figure(figsize=(8,8))
sns.scatterplot(x = 'sepal_length', y = 'sepal_width', data = iris_df, hue = 'species', size='species')
plt.show() 
```

```python
plt.figure(figsize=(12,8))
sns.boxplot(data = iris_df, palette='magma')
```
